```python

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import numpy as np

# Etiketler Ã¼zerinde LabelEncoder kullanÄ±mÄ±
le = LabelEncoder()
ySample = le.fit_transform(ySample)  # Etiketleri sayÄ±sal hale getir

np.save('classes.npy',le.classes_)  # Etiketleri kaydet
print("SÄ±nÄ±flar kaydedildi:", le.classes_)

# Ã–zellikler Ã¼zerinde MinMaxScaler kullanÄ±mÄ±
scaler = MinMaxScaler()
XSample_normalized = scaler.fit_transform(XSample)  # Ã–zellikleri normalize et

# Normalize edilmiÅŸ Ã¶zellikleri DataFrame olarak geri dÃ¶ndÃ¼r
XSample_normalized = pd.DataFrame(XSample_normalized, columns=XSample.columns)

print(XSample_normalized.head())  # Ä°lk 5 normalize edilmiÅŸ Ã¶zellik
print(ySample[:10])  # Ä°lk 10 etiketin sayÄ±sal hali
```

```python
from sklearn.preprocessing import MinMaxScaler
import joblib

# Scaler oluÅŸturuluyor ve veri Ã¼zerinde uygulanÄ±yor
scaler = MinMaxScaler()
XSample_normalized = scaler.fit_transform(XSample)

# Scaler'Ä± kaydet
joblib.dump(scaler, "Scaler.pkl")
print("Scaler kaydedildi.")
```

```python
from sklearn.model_selection import train_test_split

XTrain,XTest,yTrain,yTest=train_test_split(XSample,ySample,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=10, metric="minkowski",p=5)
model.fit(XTrain,yTrain)

yTestPred = model.predict(XTest)#modele testte bulun diyoruz.


```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.utils.multiclass import unique_labels

# Classification report yazdÄ±rma
print(classification_report(yTest, yTestPred))

# Benzersiz sÄ±nÄ±flarÄ± dinamik olarak al
labels = unique_labels(yTest, yTestPred)

# Confusion matrix hesaplama ve normalize etme
cm = confusion_matrix(yTest, yTestPred, labels=labels, normalize="true").round(2)

# Confusion matrix gÃ¶rselleÅŸtirme
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp = disp.plot(cmap=plt.cm.GnBu, values_format='.3f')

# BaÅŸlÄ±k ekleme
plt.title(f"Train Confusion Matrix", style="italic")

# GÃ¶rselleÅŸtirmeyi gÃ¶ster
plt.show()
```
