```python

from sklearn.preprocessing import MinMaxScaler, LabelEncoder
import numpy as np

# Etiketler üzerinde LabelEncoder kullanımı
le = LabelEncoder()
ySample = le.fit_transform(ySample)  # Etiketleri sayısal hale getir

np.save('classes.npy',le.classes_)  # Etiketleri kaydet
print("Sınıflar kaydedildi:", le.classes_)

# Özellikler üzerinde MinMaxScaler kullanımı
scaler = MinMaxScaler()
XSample_normalized = scaler.fit_transform(XSample)  # Özellikleri normalize et

# Normalize edilmiş özellikleri DataFrame olarak geri döndür
XSample_normalized = pd.DataFrame(XSample_normalized, columns=XSample.columns)

print(XSample_normalized.head())  # İlk 5 normalize edilmiş özellik
print(ySample[:10])  # İlk 10 etiketin sayısal hali
```

```python
from sklearn.preprocessing import MinMaxScaler
import joblib

# Scaler oluşturuluyor ve veri üzerinde uygulanıyor
scaler = MinMaxScaler()
XSample_normalized = scaler.fit_transform(XSample)

# Scaler'ı kaydet
joblib.dump(scaler, "Scaler.pkl")
print("Scaler kaydedildi.")
```

```python
from sklearn.model_selection import train_test_split

XTrain,XTest,yTrain,yTest=train_test_split(XSample,ySample,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=10, metric="minkowski",p=5)
model.fit(XTrain,yTrain)

yTestPred = model.predict(XTest)#modele testte bulun diyoruz.


```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.utils.multiclass import unique_labels

# Classification report yazdırma
print(classification_report(yTest, yTestPred))

# Benzersiz sınıfları dinamik olarak al
labels = unique_labels(yTest, yTestPred)

# Confusion matrix hesaplama ve normalize etme
cm = confusion_matrix(yTest, yTestPred, labels=labels, normalize="true").round(2)

# Confusion matrix görselleştirme
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp = disp.plot(cmap=plt.cm.GnBu, values_format='.3f')

# Başlık ekleme
plt.title(f"Train Confusion Matrix", style="italic")

# Görselleştirmeyi göster
plt.show()
```
