```python
import pandas as pd

dataset = pd.read_csv(r'C:\VoiceProject\Data\mfcc_features_labels.csv')

dataset
```

```python
dataset.info()
```

```python
XSample = dataset.iloc[:,:98]
ySample = dataset.Kisi

XSample
ySample
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

ySample = le.fit_transform(ySample)
ySampleName = le.classes_

print(ySampleName)
print(ySample)
```

```python
for item in dataset.columns[:98]:
    XSample[item] = le.fit_transform(XSample[item])
    
print(XSample)
```

```python
print(XSample.corr())
```

```python
threshold = 0.7
lower_threshold = -0.7  # Negatif eşik değeri

# Hem pozitif hem negatif eşik değerlerini kontrol edelim
filtered_corr = XSample.corr().where(
    (XSample.corr() > threshold) | (XSample.corr() < lower_threshold)
)

# Korelasyonu yüksek olan özellik çiftlerini yazdıralım
for column in filtered_corr.columns:
    correlated_columns = filtered_corr[column].dropna().index
    if len(correlated_columns) > 0:
        print(f"Korelasyonu yüksek özellikler: {column} ve {', '.join(correlated_columns)}")
```

```python
threshold = 0.7
lower_threshold = -0.7  # Negatif eşik değeri

# Korelasyon matrisini al ve eşik değerlere göre filtrele
correlation_matrix = XSample.corr()
filtered_corr = correlation_matrix.where(
    (correlation_matrix > threshold) | (correlation_matrix < lower_threshold)
)

# En az 5 yüksek korelasyonu olan özellikleri bul
columns_to_drop = []
for column in filtered_corr.columns:
    correlated_columns = filtered_corr[column].dropna().index
    if len(correlated_columns) >= 5:  # Korelasyonu yüksek olan 5 veya daha fazla
        columns_to_drop.append(column)

# Tekrarlardan kaçınmak için benzersiz kolonları al
columns_to_drop = list(set(columns_to_drop))

# Bulunan kolonları XSample'dan sil
XSample.drop(columns=columns_to_drop, inplace=True)

print(f"Silinen kolonlar: {columns_to_drop}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Korelasyon matrisini hesapla
correlation_matrix = XSample.corr()

# Isı haritası için bir grafik boyutu ayarla
plt.figure(figsize=(12, 10))

# Seaborn heatmap ile korelasyonları görselleştir
sns.heatmap(
    correlation_matrix, 
    annot=True,  # Hücrelere sayısal değerleri ekle
    fmt=".2f",  # Sayı formatı
    cmap="coolwarm",  # Renk skalası
    cbar=True,  # Sağ tarafta renk skalası göster
    square=True  # Hücrelerin kare görünmesini sağlar
)

# Başlık ve göster
plt.title("Korelasyon Isı Haritası", fontsize=16)
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

# MinMaxScaler oluştur
scaler = MinMaxScaler()

# XSample değerlerini normalize et
XSample_normalized = scaler.fit_transform(XSample)

# Normalize edilmiş değerleri DataFrame olarak geri döndür
XSample_normalized = pd.DataFrame(XSample_normalized, columns=XSample.columns)

print(XSample_normalized.head())
```

```python
print(XSample.columns)
XSample.columns.__len__()

```

```python
from sklearn.model_selection import train_test_split

XTrain,XTest,yTrain,yTest=train_test_split(XSample,ySample,test_size=0.20,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=11, metric="minkowski",p=4)
model.fit(XTrain,yTrain)

yTestPred = model.predict(XTest)#modele testte bulun diyoruz.


```

```python
from sklearn.metrics import classification_report, ConfusionMatrixDisplay, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.utils.multiclass import unique_labels

# Classification report yazdırma
print(classification_report(yTest, yTestPred))

# Benzersiz sınıfları dinamik olarak al
labels = unique_labels(yTest, yTestPred)

# Confusion matrix hesaplama ve normalize etme
cm = confusion_matrix(yTest, yTestPred, labels=labels, normalize="true").round(2)

# Confusion matrix görselleştirme
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp = disp.plot(cmap=plt.cm.GnBu, values_format='.3f')

# Başlık ekleme
plt.title(f"Train Confusion Matrix", style="italic")

# Görselleştirmeyi göster
plt.show()
```
